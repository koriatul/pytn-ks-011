**Simple Linear Regression**

```python
x = np.array([5,15,25,35,45,55]).reshape(-1,1)
y = np.array([5,20,14,32,22,38])
print(x)
print(y)
```

```python
model = LinearRegression()
```

```python
model.fit(x,y)
```

```python
# R-Squared
r_sq = model.score(x,y)
r_sq
```

```python
print('intercept:', model.intercept_)
print('slope:' , model.coef_)
```

y = mx + b

y adalah target

m adalah slope

b adalah intercept

```python
print('Formula dari model adalah y = ', model.coef_[0], 'x + ', model.intercept_)
```

```python
y_pred = model.predict(x)
y_pred
```

```python
print(x)
print(y)
```

```python
plt.scatter(x,y, alpha=0.5)
plt.plot(x,y_pred)

plt.title('Scatter plot x and y')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
x_new = np.arange(5,20,5).reshape([-1,1])
print(x_new)
```

```python
y_new_pred = model.predict(x_new)
print(x_new)
```

**Advanced Linear Regression with Statsmodels**

```python
import statsmodels.api as sm
```

```python
x = np.array([[0,1],
             [5,1],
             [15,2],
             [25,5],
             [35,11],
             [45,15],
             [55,34],
             [60,35]])
y = np.array([4, 5, 20, 14, 32, 22, 38, 43])
```

```python
x
```

```python
model_stat = sm.OLS(y,x)
```

```python
results = model_stat.fit()
```

```python
print(results.summary())
```

```python
print('rsquared:',results.rsquared)
```

```python
results.params
```

```python
res_summary = results.summary()
res_summary.tables[1]
```

```python
res_summary.tables[0]
```

```python
y_pred = results.predict(x)
```

```python
mse = mean_squared_error(y,y_pred)
```

```python
np.sqrt(mse)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
linreg = LinearRegression()
linreg.fit(X_train,y_train)
```

```python
linreg.score(X_train,y_train)
```

```python
y_pred = linreg.predict(X_test)
```

```python
plt.scatter(X_test,y_test)
plt.plot(X_test,y_pred)
```

```python
y_pred_train = linreg.predict(X_train)
y_pred_test = linreg.predict(X_test)
```

```python
mse_train = mean_squared_error(y_train,y_pred_train)
mse_test = mean_squared_error(y_test,y_pred_test)
rmse_tr = np.sqrt(mse_train)
rmse_test = np.sqrt(mse_test)
print('train mse:',mse_train)
print('test mse',mse_test)
print('train rmse',rmse_tr)
print('test rmse',rmse_test)
```
