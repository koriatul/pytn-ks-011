```python
from sklearn.datasets import load_breast_cancer
```

```python
df_breast = load_breast_cancer()
```

```python
df_breast.feature_names
```

```python
df_breast.target_names
```

```python
import pandas as pd
```

```python
#membuat data frame
df = pd.DataFrame(df_breast.data,columns=df_breast.feature_names)
df
```

```python
df['target'] = df_breast.target
```

```python
df
```

```python
df['target'].replace(0,'Benign',inplace=True)
df['target'].replace(1,'Malignant',inplace=True)
```

```python
df.target.value_counts()
```

```python
from sklearn.preprocessing import StandardScaler
x = df.drop('target',axis=1)
stdscale = StandardScaler()
x_scaled = stdscale.fit_transform(x)
x_scaled
```

```python
df_feature_scaled = pd.DataFrame(x_scaled,columns=df_breast.feature_names)
df_feature_scaled
```

<h2> PCA sklearn <h2>

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
features_pca = pca.fit_transform(x_scaled)
features_pca
```

```python
pca_df = pd.DataFrame(features_pca,columns=['PC1','PC2'])
pca_df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(20,20))
sns.heatmap(df_feature_scaled.corr(),annot=True)
```

Data banyak multikolinear

```python
print('Explained Variation per Principal Component', pca.explained_variance_ratio_)
```

pada PC1 menampung 44.2% dari informasi data sedangkan PC2 menampung 18.9% dari data sehingga ada 36% information loss

<h3> Visualisasi data setelah PCA <h3>

```python
import matplotlib as mpl
import matplotlib.pyplot as plt

%matplotlib inline

plt.figure(figsize=(10,10))
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
targets=['Benign','Malignant']
colors  =['r','g']
for target,color in zip(targets,colors):
    indicesToKeep = df['target'] ==target
    plt.scatter(pca_df.loc[indicesToKeep,'PC1'],
               pca_df.loc[indicesToKeep,'PC2'],c=color,s=40)
plt.legend(targets)
plt.title('PCA of Breast Cancer Data')
plt.xlabel('PC1')
plt.ylabel('PC2')

plt.show()
```

<h3> Kmeans with SKlearn <h3>

```python
from sklearn.cluster import KMeans
import numpy as np
```

```python
X = -2* np.random.rand(100,2)
X1 = 1 +2 * np.random.rand(50,2)
X[50:100,:] = X1

plt.scatter(X[:,0],X[:,1],s=40,c='b')
plt.show()
```

```python
import numpy
numpy.__version__
```

```python
np.__version__
```

```python
kmeans = KMeans(n_clusters=2)
kmeans.fit(X)
```

```python
inertia = []
for i in range(1,11):
    kmeans = KMeans(n_clusters=i,init='k-means++',max_iter=300,n_init=10,random_state=0)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)
    
plt.plot(range(1,11),inertia)
plt.title('Elbow Method')
plt.xlabel('No. of Clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
kmeans_best = KMeans(n_clusters=2)
kmeans_best.fit(X)
```

```python
kmeans_best.cluster_centers_
```

```python
plt.scatter(X[:,0],X[:,1],s=40,c='b')
plt.scatter(1.93640656,1.99708516,s=40,c='r')
plt.scatter(-0.99600977,-0.96411206,s=40,c='green')
```

<h3> KMeans dengan real data <h3>

```python
import pandas as pd
import matplotlib.pyplot
import seaborn as sns
import numpy as np
from sklearn.cluster import KMeans
%matplotlib inline
```

```python
df_wholesale = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/00292/Wholesale%20customers%20data.csv')
```

```python
df_wholesale
```

```python
df_wholesale.info()
```

```python
df_wholesale.describe()
```

```python
# Standardzing the data
from sklearn.preprocessing import StandardScaler
stdscale = StandardScaler()
wholesale_scaled = stdscale.fit_transform(df_wholesale)

pd.DataFrame(wholesale_scaled,columns=df_wholesale.columns).describe()
```

```python
kmeans_2  = KMeans(n_clusters=2,init='k-means++')

kmeans_2.fit(wholesale_scaled)
```

```python
kmeans_2.inertia_
```

```python
inertia =[]
for k in range(1,25):
    kmeans = KMeans(n_clusters=k,init='k-means++')
    kmeans.fit(wholesale_scaled)
    inertia.append(kmeans.inertia_)

plt.plot(range(1,25),inertia)
plt.title('elbow method')
plt.xlabel('Jumlah Clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
kmeans_6  = KMeans(n_clusters=6,init='k-means++')
kmeans_6.fit(wholesale_scaled)
pred = kmeans_6.predict(wholesale_scaled)
```

```python
df_scaled = pd.DataFrame(wholesale_scaled,columns=df_wholesale.columns)
df_scaled
```

```python
df_scaled['clusters'] = pred
df_scaled
```

```python
df_scaled.clusters.value_counts()
```

```python
df_wholesale['clusters'] = pred
df_wholesale
```

```python
df_wholesale.clusters.value_counts()
```

```python
df_wholesale[df_wholesale.clusters==2].describe()
```
