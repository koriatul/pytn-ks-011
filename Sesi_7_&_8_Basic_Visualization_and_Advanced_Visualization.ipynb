```python
#Mapping 100 row pertama dari tabel insiden
df_incidents_100 = df_incidents.iloc[0:100,:]
df_incidents_100
```

```python
#San Fransisco Lat and Lon
latitude= 37.77
longitude=-122.42
```

```python
sf_map = folium.Map(location=[latitude,longitude],zoom_start=12)
sf_map
```

```python
img = Image.open('/content/drive/MyDrive/Colab Notebooks/gl and gb.png')
img.show()
```

```python
df_incidents.head()
```

```python
df_incidents_100
```

```python
sf_map = folium.Map(location=[latitude,longitude],zoom_start=12)
sf_map
incidents = folium.map.FeatureGroup()

# add 100 crimes to FeatureGroup

for lat,lon in zip(df_incidents_100['Y'],df_incidents_100['X']):
    incidents.add_child(
        folium.CircleMarker(
            [lat,lon],
            radius=5,
            color='blue',
            fill=True,
            fill_color='red',
            fill_opacity=0.6
        
        )
    
    )
sf_map.add_child(incidents)
```

```python
df_incidents_100.Category
```

```python
sf_map = folium.Map(location=[latitude,longitude],zoom_start=12)
sf_map
incidents = folium.map.FeatureGroup()

# add 100 crimes to FeatureGroup

for lat,lon,label in zip(df_incidents_100['Y'],df_incidents_100['X'],df_incidents_100.Category):
    incidents.add_child(
        folium.CircleMarker(
            [lat,lon],
            radius=5,
            color='blue',
            fill=True,
            fill_color='red',
            fill_opacity=0.6,
            popup=label
        
        )
    
    )
sf_map.add_child(incidents)
```

```python
# Clustering semua data sesuai kemiripan
from folium import plugins
sf_map = folium.Map(location=[latitude,longitude],zoom_start=12)

incidents = plugins.MarkerCluster().add_to(sf_map)

for lat,lon,label in zip(df_incidents_100['Y'],df_incidents_100['X'],df_incidents_100.Category):
    incidents.add_child(
        folium.CircleMarker(
            [lat,lon],
            radius=5,
            color='blue',
            fill=True,
            fill_color='red',
            fill_opacity=0.6,
            popup=label
        
        )
    
    )
```
